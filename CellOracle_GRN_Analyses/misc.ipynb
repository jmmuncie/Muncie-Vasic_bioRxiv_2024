```python
import pandas as pd
```

```python
timepoint = 85
```

```python
base_GRN = pd.read_parquet(f'data/base_grn_outputs/E{timepoint}/base_GRN_dataframe.parquet')
```

```python
diffexp_wt_up = pd.read_csv(
    f'./data/base_grn_outputs/E{timepoint}/E{timepoint}_diff_access_wt.txt'
    , sep = '\t')
diffexp_ko_up = pd.read_csv(
    f'./data/base_grn_outputs/E{timepoint}/E{timepoint}_diff_access_ko.txt'
    , sep = '\t')
```

```python
diffexp_ko_up
```

```python
peaks_only_in_ko = []
for index, row in diffexp_ko_up.iterrows():
    peaks_only_in_ko.append(f"{row['seqnames']}_{row['start']}_{row['end']}")
    
peaks_only_in_wt = []
for index, row in diffexp_wt_up.iterrows():
    peaks_only_in_wt.append(f"{row['seqnames']}_{row['start']}_{row['end']}")
```

```python
ko_peaks = base_GRN[base_GRN.peak_id.isin(peaks_only_in_ko)]
ko_peaks[ko_peaks.loc[:, 'Nr2f2'] == 1]
```

```python
wt_peaks = base_GRN[base_GRN.peak_id.isin(peaks_only_in_wt)]
wt_peaks[wt_peaks.loc[:, 'Nr2f2'] == 1]
```
